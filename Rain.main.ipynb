Rainfall in Iran analyse


import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Read Dataset > Rainfall monthly dataset from crudata monthly from 1901-2022

```python
dataframe = pd.read_csv("Rainfall.csv")
dataframe[:5]
```

Analyse types

```python
dataframe.info()
dataframe
```

Convert date to month

```python
df = pd.DataFrame(dataframe)
df['date'] = pd.to_datetime(df['date'])
df['date'] = df['date'].dt.strftime('%b')
df
```

```python
dataframe.describe().T
```

Using melt to Hvae specific data

```python
dataframe['date'] = pd.to_datetime(dataframe['date'])

data_melt=pd.melt(dataframe, id_vars=['date'], var_name='city', value_name='rainfall')
data_melt.head(15)
```

Cheking isnull

```python
data_melt.isnull().sum()
```

boxploting to Check overall data

```python
plt.figure(figsize=(16, 20))
sns.boxplot(data=data_melt,x='city',y='rainfall')
plt.xticks(rotation=90)
```

According to the boxplot, there are some cities with rainfall values larger than usual like Ilam

```python
unusual_rainfall=data_melt[data_melt['rainfall']>350]
unusual_rainfall['city'].value_counts()
```

As we can see there are 360 unusuall valuse in Ilam data so we have to delete this city 

```python
data_melt=data_melt[data_melt['city']!='Ilam']
data_melt
```

Now we Add 4 more columns
Season
Decade
Year
Month

```python
#Decade
data_melt['decade']=((data_melt['date'].dt.year.astype(int)-1900)//10).astype(int)
data_melt['decade']=(1900+10*data_melt['decade']).astype(str)+'-'+(1909+10*data_melt['decade']).astype(str)
data_melt[['date','decade']]
```

```python
#Year
data_melt['year']=data_melt['date'].dt.year
data_melt[['date','year']]
```

```python
#Season
def get_season(month):
    if 3 <= month <= 5:
        return 'Spring'
    elif 6 <= month <= 8:
        return 'Summer'
    elif 9 <= month <= 11:
        return 'Fall'
    else:
        return 'Winter'
    
data_melt['season']=data_melt['date'].dt.month.apply(get_season)
data_melt[['date','season']]
```

```python
#Month
data_melt['month']=data_melt['date'].dt.month
data_melt[['date','month']]
```

```python
data_melt.head()
```

Analysing the years

```python
yearly_mean= pd.pivot_table(data_melt, values='rainfall', index='year', aggfunc='mean')

sns.regplot(data=yearly_mean,x=yearly_mean.index,y='rainfall')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.lineplot(data=yearly_mean,x=yearly_mean.index,y='rainfall')
plt.xticks(rotation=90)
plt.show()
```

According to these charts there is no relation btw rainfall and years

```python
#Droping city
df2 = data_melt.drop(['city','decade','season','date'], axis=1)


sns.heatmap(df2.corr(),annot=True, annot_kws={"size": 13},cbar=False)
```

Analysing by seasons

```python
season_mean= pd.pivot_table(data_melt, values='rainfall', index='city', columns='season', aggfunc='mean', margins=True, margins_name='AVERAGE')
season_mean
```

Analysing average rainfall by season

```python
season_mean=season_mean.iloc[:,:-1]

seasons_order = ['Spring', 'Summer', 'Fall', 'Winter']

season_res = season_mean.reindex(columns=seasons_order)

sns.barplot(x=season_res.columns, y=season_res.loc['AVERAGE'])
plt.title('Average Rainfall by Season')
plt.ylabel('Average Rainfall')
plt.xlabel('Season')

plt.show
```

```python
# remove Average row
season_mean=season_mean.iloc[:-1,:]
season_mean
```

Now we plot Average rainfall by season for cities

```python
cities = season_mean.index

num_rows = 8
num_cols = 4  

total_subplots = 30

fig, axes = plt.subplots(num_rows, num_cols, figsize=(16, 4*num_rows))

axes = axes.flatten()
custom_order = ['Spring', 'Summer', 'Fall', 'Winter']

color = [ 'blue', 'purple', 'red', 'black']

for i, city in enumerate(cities):
    city_data = season_mean.loc[city]
#     print(city_data)
    ax = axes[i]
    sns.barplot(x=city_data.index,y=city_data.values,order=custom_order, ax=ax,color = 'blue', palette = ['tab:blue', 'tab:orange', 'tab:red', 'tab:green'])
    ax.set_title(f'Rainfall by Season in {city}')
    ax.set_ylim(0, 140)
        
    ax.set_xlabel('')
    ax.set_label('Rainfall')
    if (i% num_cols!=0):
        ax.set_ylabel('')
    
for j in range(total_subplots, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()

# Show the plot
plt.show()
```

Now we checking Decades


```python

decade_mean= pd.pivot_table(data_melt, values='rainfall', index='city', columns='decade', aggfunc='mean', margins=True, margins_name='AVERAGE')
decade_mean
```

```python
# remove Average column

decade_mean=decade_mean.iloc[:,:-1] 

sns.lineplot(x=decade_mean.columns, y=decade_mean.loc['AVERAGE'])
plt.title('Average Rainfall by decade')
plt.ylabel('Average Rainfall')
plt.xlabel('Decade')
plt.ylim(23, 30) 
plt.xticks(rotation=90)

plt.show()
```

The above chart shows that rainfall is decreasing through time passes


Monthly mean

now we analysing The monthly mean

```python
#months
month_mean= pd.pivot_table(data_melt, values='rainfall', index='city', columns='month', aggfunc='mean', margins=True, margins_name='AVERAGE')
month_mean
```

```python
# remove Average column
month_mean=month_mean.iloc[:,:-1] 

sns.lineplot(x=month_mean.columns, y=month_mean.loc['AVERAGE'])
plt.title('Average Rainfall by month')
plt.ylabel('Average Rainfall')
plt.xlabel('Month')

month_names = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
plt.xticks( range(1,13),month_names, rotation=90)

plt.show()
```

According to this plot, the average rainfall in Iran is max in March and decreases  in July and August

```python
# remove Average row
month_mean=month_mean.iloc[:-1,:]
```

Month mean of all cities


```python
cities = month_mean.index

num_rows = 10
num_cols = 3  


total_subplots = 30

# Create subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 4*num_rows))


axes = axes.flatten()


for i, city in enumerate(cities):
    city_data = month_mean.loc[city]
    ax = axes[i]
    sns.lineplot(data=city_data, ax=ax, errorbar=None,color="red")
    ax.set_title(f'Rainfall by month in {city}')

    ax.set_ylabel('Rainfall')
    
    ax.set_xlabel('')
    if (i% num_cols!=0):
        ax.set_ylabel('')
    
    month_names=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    ax.set_xticks(range(1, 13))
    ax.set_xticklabels(month_names, rotation=90)
    
    ax.set_ylim(0, 145) 

# Remove any unused subplots
for j in range(total_subplots, len(axes)):
    fig.delaxes(axes[j])


plt.tight_layout()


plt.show()
```
